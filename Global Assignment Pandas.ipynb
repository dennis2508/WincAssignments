```python
import pandas as pd
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
all_countries = tables[1]
all_countries
```

```python
all_countries.columns = ["Country","1990", "2005", "2017", "2017%World", "2017 vs 1990 Change", "Land Area", "Per Capita", "Total in LUCF", "Total ex LUCF"]
countries = all_countries.loc[4:]
countries = countries.drop([67])
countries
```

```python
top_5 = countries.sort_values(by = ["2017"], ascending = False).head(5)
top_5 = top_5.loc[:,["Country","1990","2005","2017"]]
top_5
```

```python
import matplotlib.pyplot as plt
import numpy as np
years = top_5.columns[1:]
years

for index, row in top_5.iterrows():
  plt.plot(years, row[1:], label=row[0])

plt.title("CO2 emmission of the 5 biggest countries")
plt.xlabel("Year")
plt.ylabel("Mt C02 per year")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()
```

```python
countries_changes = countries.loc[:,["Country","1990","2005","2017"]]
countries_changes["Basis = 1990"] = 100
countries_changes["1990-2005"] = (countries_changes["1990"] / countries_changes["2005"])*100
countries_changes["1990-2017"] = (countries_changes["1990"] / countries_changes["2017"])*100
countries_changes_top_3 = countries_changes.sort_values(by =["1990-2017"], ascending=True).head(3)
countries_changes_bottom_3 = countries_changes.sort_values(by = ["1990-2017"], ascending=False).head(3)
top_bottom = countries_changes_top_3.append(countries_changes_bottom_3)
top_bottom
```

```python
years = top_bottom.columns[1:4]
years

for index, row in top_bottom.iterrows():
  plt.plot(years, row[4:], label=row[0])

plt.title("Worst and best changers")
plt.xlabel("Year")
plt.ylabel("%")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()
```

```python
countries_changes.columns = ["Country", "Start1990", "2005", "2017", "Basis = 1990", "1990-2005", "1990-2017"]
countries_changes_5_least = countries_changes.loc[countries_changes.Start1990 >= 5]
countries_changes_5_top_3 = countries_changes_5_least.sort_values(by =["1990-2017"], ascending=True).head(3)
countries_changes_5_bottom_3 = countries_changes_5_least.sort_values(by = ["1990-2017"], ascending=False).head(3)
top_bottom_5_least = countries_changes_5_top_3.append(countries_changes_5_bottom_3)
top_bottom_5_least
```

```python
years = top_bottom_5_least.columns[1:4]
years

for index, row in top_bottom_5_least.iterrows():
  plt.plot(years, row[4:], label=row[0])

plt.title("Worst and best changers")
plt.xlabel("Year")
plt.ylabel("%")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()
```
